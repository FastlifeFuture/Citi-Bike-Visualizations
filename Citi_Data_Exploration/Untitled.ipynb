```python
#import pandas to read csv's & create/manipulate dataframes
import pandas as pd

citi_data = pd.read_csv("../Citi_bike_data/tripdata.csv")
citi_data1 = pd.read_csv("../Citi_bike_data/tripdata1.csv")
citi_data2 = pd.read_csv("../Citi_bike_data/tripdata2.csv")
citi_data3 = pd.read_csv("../Citi_bike_data/tripdata3.csv")

#concatenate the data into 1 data frame
frames = [citi_data, citi_data1, citi_data2, citi_data3]
citi_bike_data = pd.concat(frames )

#write the dataframe into a csv
citi_bike_data.to_csv("../citi_bike_data/citi_bike_data.csv")
```
